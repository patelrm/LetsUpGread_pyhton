```python
#[Q1].prime number 1  to 1000 usinf for loop
for num in range(1,1001):
    count=0
    for i in range(2,(num//2+1)):
        if(num%i==0):
            count=count+1
            break
if(count==0 and num!=1):
    print("%d" %num ,end=' ')
```

```python
#[Q2.]multiplication table of 1 to 10 useing for loop
for i in range(1,11):
    print("\n \n MULTIPLICATION TABLE %d\n " %(i))
    for j in range(1,11):
        print("%3d\t X %3d = %3d"%(i,j,i*j))
```
